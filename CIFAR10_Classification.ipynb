```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
x_train[0].shape
```

```python
plt.imshow(x_train[0])
```

```python
plt.imshow(x_train[30])
```

```python
x_train = x_train/255
```

```python
x_test = x_test/255
```

```python
x_test.shape
```

```python
y_train
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_cat_train = to_categorical(y_train,10)
y_cat_test = to_categorical(y_test,10)
```

```python
y_cat_train[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=32,kernel_size=(4,4),padding='valid',input_shape=(32,32,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters=32,kernel_size=(4,4),padding='valid',input_shape=(32,32,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

#flattening the images
model.add(Flatten())
model.add(Dense(256,activation='relu'))

#OUTPUT Layer --> Softmax
model.add(Dense(10,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
model.fit(x_train,y_cat_train,epochs=15,validation_data=(x_test,y_cat_test),
          callbacks=[early_stop])
```

```python
model.save('cifar10_keras.h5')
```

```python
metrics = pd.DataFrame(model.history.history)
metrics[['loss','val_loss']].plot()
```

```python
metrics[['accuracy','val_accuracy']].plot()
```

```python
model.evaluate(x_test,y_cat_test,verbose=0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predict_x=model.predict(x_test) 
classes_x=np.argmax(predict_x,axis=1)
```

```python
print(classification_report(y_test,classes_x))
```

```python
confusion_matrix(y_test,classes_x)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test,classes_x),annot=True)
```
